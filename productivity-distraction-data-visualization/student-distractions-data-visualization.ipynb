# Visualizing Student Productivity and Distractions through Lifestyle and Academic Variables

Specifically, a Pew Research article by Hatfield (2024) shared that over 50% of U.S. teachers for high school students reported that a significant disruption to classroom learning was phone use.
This dataset, *Student Productivity & Digital Distraction Dataset*, was created by Preet (2026) and was sourced from Kaggle. This dataset includes many variables reported about student lifestyle and academics, such as hours of studying, sleeping, and phone usage, as well as variables such as stress, productivity, and final grade (Preet, 2026). Through data cleaning and analysis, visualizations that compare these variables can provide evidence and draw conclusions regarding distractions and academic productivity in students.

Methods for this data visualization analysis include loading in the necessary packages and the dataset, identifying the format of the data and data types of the columns, cleaning the dataset, deriving summary statistics, comparing variables through data manipulation, re-formatting the data to long-format, and using python visualization packages to create several plots that describe the overall trends in the dataset.

## Data Analysis

To begin, the packages needed for this analysis, `pandas`, `matplotlib`, and `seaborn` need to be imported.

```python
# Import pandas
import pandas as pd
# Import matplotlib
import matplotlib.pyplot as plt
# Import seaborn
import seaborn as sns
```

The dataset, *student_productivity_distraction_dataset_20000.csv*, will be read in and information regarding the data types and shape can be derived.

```python
# Read in the data as a new variable `student_data`
student_data = pd.read_csv("student_productivity_distraction_dataset_20000.csv")
# Check the data types
student_data.info()
```

```python
# Get information on the data shape
print(student_data.shape)
```

This dataset has 20,000 rows and 18 columns. All of the data types are consistent and do not need to be cleaned for this analysis.

Before exploring summary statistics, the data will be checked and cleaned for NA values and duplicates.

```python
# Check for NA values
student_data.isna().sum()
```

```python
# Check for duplicate values
student_data.duplicated().sum()
```

The dataset is already nicely cleaned, no NA values or duplicates are present, and the data types are all correct and do not need to be stored differently. Now, summary statistics can provide an overview of each column of data.

```python
# Derive summary statistics
student_data.describe()
```

## Data Manipulation and Analysis

First, using some data frame manipulations, variables can be compared to get a better sense of the trends within the dataset.

Compare the minimum study hours (0.5) with the maximum study hours (10.00) and see how productivity score varies between the two.

```python
# Pull the columns `study_hours_per_day` and `productivity_score` and save as a new variable
study_and_productivity_low = student_data[["study_hours_per_day", "productivity_score"]]
# Filter the new variable to only contain the minimum hours studied (0.50) in `study_hours_per_day`
study_low = study_and_productivity_low[(study_and_productivity_low["study_hours_per_day"] == 0.500000)]
# Derive summary statistics for this new variable
study_low.describe()
```

```python
# Pull the columns `study_hours_per_day` and `productivity_score` and save as a new variable
study_and_productivity_high = student_data[["study_hours_per_day", "productivity_score"]]
# Filter the new variable to only contain the maximum hours studied (10.00) in `study_hours_per_day`
study_high = study_and_productivity_high[(study_and_productivity_high["study_hours_per_day"] == 10.000000)]
# Derive summary statistics for this new variable
study_high.describe()
```

When comparing the minimum study hours, maximum study hours, and calculating summary statistics, the mean for `productivity_score` is much higher with maximum study hours than with minimum.

Next, average phone usage with gender can be compared.

```python
# Use groupby() to take the mean of `phone_usage_hours` by `gender`
student_data.groupby("gender")["phone_usage_hours"].mean()
```

On average, each gender has around the same amount of phone usage hours, ~6.2 hours.

Next, average `productivity_score` based on each `study_hours_per_day` and `phone_usage_hours`.

```python
# Use groupby() to take the mean `productivity_score` per study and phone hours
student_data.groupby(["study_hours_per_day", "phone_usage_hours"], as_index = False)["productivity_score"].mean()
```

From this overview, it is hypothesized that with both minimum and maximum studying, increasing phone usage decreases productivity. However, visualizing this relationship can offer better insights into this trend.

```python
# Make a new variable that pulls `study_hours_per_day`, `phone_usage_hours`, and `productivity_score` and will hold the max study hours
study_ten = student_data[["study_hours_per_day", "phone_usage_hours", "productivity_score"]]
# Make a new variable that contains only the maximum hours (10.00) of studying
study_ten_separate = study_ten[(study_ten["study_hours_per_day"] == 10.000000)]
# Make a simple plot that takes the mean `productivity_score` by the max study hours and increasing phone usage
ten_plot = study_ten_separate.groupby(["study_hours_per_day","phone_usage_hours"], as_index = False)["productivity_score"].mean()
ax = ten_plot.plot(kind="line")
ax.set_xlabel("Hours")
ax.set_ylabel("Score")
ax.set_title("Comparing Increased Phone Usage and Maximum Studying with Productivity")
plt.show()
```

From this quick plot, it is noted that as `phone_usage_hours` increases, `productivity_score` decreases. However, a stronger visualization plotting the direct relationship between the two, `phone_usage_hours` and `productivity_score`, can be created to identify the average trend over the entire data, not just with maximum study hours.

```python
# Save a new variable that uses groupby() to take the average `productivity_score` per `phone_usage_hours`
phone_productivity_plot = student_data.groupby("phone_usage_hours")["productivity_score"].mean()
# Plot this relationship as a line plot
ax1 = phone_productivity_plot.plot(kind="line")
ax1.set_xlabel("Phone Usage Hours")
ax1.set_ylabel("Productivity Score (0-100)")
ax1.set_title("Average Productivity Score across Phone Usage")
plt.show()
```

Looking at the overall trend from the entire dataset, `productivity_score` on average has a negative association with `phone_usage_hours`. Increasing phone use decreases productivity score.

## Data Formatting
For two of the visualizations created below, a few of the columns from the dataset will need to be reformatted into long format from wide format. This section will carry out these data manipulations.

```python
# Make a new variable that will combine `phone_usage_hours` and `study_hours_per_day` into one column named `Phone_Usage_and_Studying` and save their values as `hours`
student_data_long_hours = student_data.melt(id_vars="productivity_score", value_vars=["phone_usage_hours","study_hours_per_day"], var_name="Phone_Usage_and_Studying", value_name="hours")
print(student_data_long_hours)
```

```python
# Make a new variable that will combine `social_media_hours`, `youtube_hours`, and `gaming_hours` into `usage_hour` and their values as `number_hour`
student_data_long_usage = student_data.melt(id_vars = "productivity_score", value_vars = ["social_media_hours", "youtube_hours", "gaming_hours"], 
                                            var_name = "usage_hour", value_name = "number_hours")
print(student_data_long_usage)
```

## Data Visualizations

## Visualization #1
This first visualization uses a line plot to compare how productivity (`productivity_score`) varies with increasing hours of phone usage versus increasing hours of studying. Using the long-format data variable created above, `student_data_long`, where phone and study hours were combined and their values placed under a new column, `hours`, a seaborn line plot can visualize the relationship between these variables and productivity.  
This dataset has 20000 rows, which can create overplotting and too much noise. To reduce the number of points on the line graph while still accurately depicting the relationship between these variables, the average `productivity_score` per `hours` was used for plotting with `student_data_long.groupby()`.

```python
# Make a new variables that averages `productivity_score` with `Phone_Usage_and_Studying` and `hours`
student_data_long_avg = student_data_long_hours.groupby(["Phone_Usage_and_Studying","hours"], as_index=False)["productivity_score"].mean()
```

```python
# Create a line plot and set the hue to define either phone usage or studying
sns.lineplot(data = student_data_long_avg, x = "hours", y = "productivity_score", hue = "Phone_Usage_and_Studying")
# Label the axes and create a title
plt.xlabel("Hours")
plt.ylabel("Productivity Score (0-100)")
plt.title("Comparison of Average Productivity Score with Increasing Phone Usage and Studying Hours")
# Create a legend to differentiate the lines
plt.legend(title = "Phone and Study Hours", title_fontsize = 9, fontsize = 8, loc = "lower center")
plt.show()
```

The results from this line plot conclude that increased hours of phone usage has a negative association with productivity, while increased study hours per day has a positive association with productivity.

## Visualization #2
This next visualization will capture the relationship between `productivity_score` and `attendance_percentage` through a scatterplot. To avoid overplotting, `productivity_score` will be averaged by `attendance_percentage`. A trend line will be added for clarity.

```python
# Make a new variable that takes the mean of `productivity_score` per `attendance_percentage`
attendance_avg = student_data.groupby("attendance_percentage")["productivity_score"].mean().reset_index()
# Create a scatterplot using the averaged data and set the alpha = 0.5
sns.scatterplot(data = attendance_avg, x = "attendance_percentage", y = "productivity_score", alpha = 0.5)
# Add a trend line for clarity and color it red
sns.regplot(data = attendance_avg, x = "attendance_percentage", y = "productivity_score", scatter=False, color = "red")
# Add axis labels and a title
plt.xlabel("Attendance Percentage")
plt.ylabel("Productivity Score (0-100)")
plt.title("Productivity Score across Percentage of Student Attendance")
plt.show()
```

While there is still quite a lot of noise in this scatterplot, there is a clear, gradual increase in productivity with increased attendance percentage. Increased attendance has a positive associaton with productivity.

## Visualization #3
This `sns.lineplot()` plots `productivity_score` with `stress_level` to identify how productivity may be affected with varying levels of stress.

```python
# Create a line plot to plot `stress_level` with `productivity_score` and set the color to red
sns.lineplot(data=student_data, x="stress_level", y="productivity_score", color = "red")
# Add axis labels and a title
plt.xlabel("Stress Level (0-10)")
plt.ylabel("Productivity Score (0-100)")
plt.title("Productivity Score with Stress Level")
plt.show()
```

Presented from the results of the line plot, there is a clear negative association with stress level and productivity.

## Visualization #4
Using the long format `student_data_long_usage` created above that combines each type of hour as `usage_hour` and their values under `number_hours`, a bar plot can be derived that visualizes the type of distraction that students spend the most time on.

```python
# Create a bar plot using the long data `student_data_long_usage` created above and set the hue to `usage_hour` to define the different value names
sns.barplot(data=student_data_long_usage, x="usage_hour", y="number_hours", hue = "usage_hour")
# Set axis labels and add a title
plt.xlabel("Usage Hour Type")
plt.ylabel("Hours")
plt.title("Hours Students Spent on Varying Distractions")
plt.show()
```

This bar plot reports that students spend the most time distracted by social media.

## Visualization #5
Lastly, two subplots were created to compare student lifestyle variables, `exercise_minutes` and `sleep_hours` with `productivity_score`. An article by the National Sleep Foundation (2025) discusses that quality amounts of sleep are vital for productivity and energy for exercise, overall supporting a healthy lifestyle. Viewing how increasing exercise and sleep affect student productivity will offer key insights and evidence for these concepts. To avoid overplotting, `productivity_score` will be averaged in both plots with `exercise_minutes` and `sleep_hours`.

```python
# Create a new variable that takes the average productivity with exercise
exercise_avg = student_data.groupby("exercise_minutes")["productivity_score"].mean().reset_index()
# Create a new variable that takes the average productivity with sleep
sleep_avg = student_data.groupby("sleep_hours")["productivity_score"].mean().reset_index()
# Create subplots
fig, ax = plt.subplots(2,1)
# Add a title above both plots
fig.suptitle("Exercise and Sleep Duration across Productivity Score")
# Create a scatterplot for the first subplot that plots exercise with productivity
# Change the color to red and set alpha = 0.5
ax[0].scatter(exercise_avg["productivity_score"], exercise_avg["exercise_minutes"], color = "red", alpha = 0.5)
# Set axis labels
ax[0].set_xlabel("Productivity Score (0-100)")
ax[0].set_ylabel("Exercise Minutes")
# Create a scatterplot for the second subplot that plots sleep with productivity, set alpha = 0.5
ax[1].scatter(sleep_avg["productivity_score"], sleep_avg["sleep_hours"], alpha = 0.5)
# Set axis labels
ax[1].set_xlabel("Productivity Score (0-100)")
ax[1].set_ylabel("Sleep Hours")
plt.tight_layout()
```

The results from these scatterplots indicate that sleep has a strong, positive association with productivity. The relationship between exercise and productivity is weak, and no association can be drawn. This plot supports the research presented by the National Sleep Foundation (2025), that both the quality and duration of sleep can greatly affect productivity and lifestyle.
